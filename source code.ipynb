```python
import numpy as np
import re
import nltk
from sklearn.datasets import load_files
nltk.download('stopwords')
nltk.download('wordnet')
import pickle
from nltk.corpus import stopwords
```

# Loading the dataset


```python
movie_data = load_files('txt_sentoken')
x,y = movie_data.data, movie_data.target
```

# Data Preprocessing

```python
documents=[]
from nltk.stem import WordNetLemmatizer
stemmer=WordNetLemmatizer() 
for sen in range(0,len(x)):
    #removing special characters
    doc=re.sub(r'\W','',str(x[sen]))
    # Removing single characters
    doc=re.sub(r'\s+[a-zA-Z]\s+', ' ', doc)
    # Remove single characters from the start
    doc=re.sub(r'\^[a-zA-Z]\s+', ' ', doc) 
    # Substituting multiple spaces with single space
    doc=re.sub(r'\s+', ' ', doc, flags=re.I)
    # Removing prefixed 'b'
    doc=re.sub(r'^b\s+', '', doc)
    # Converting to Lowercase
    doc=doc.lower()
    # Lemmatization
    doc=doc.split()

    doc=[stemmer.lemmatize(word) for word in doc]
    doc=' '.join(doc)

    documents.append(doc)
```

# convert text into number

```python
from sklearn.feature_extraction.text import TfidfVectorizer  
tfidfconverter = TfidfVectorizer(max_features=1500, min_df=5, max_df=0.9, stop_words=stopwords.words('english'))  
x = tfidfconverter.fit_transform(documents).toarray()  
```

```python
x[:6]
```

# Data Spliting into training and test set

```python
from sklearn.model_selection import train_test_split  
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)  
```

# Model Training

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators=1000, random_state=0)  
classifier.fit(x_train, y_train) 
```

```python
y_predicted=classifier.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(y_test,y_predicted)
```
